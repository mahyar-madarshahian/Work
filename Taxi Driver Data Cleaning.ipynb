```python
# import pandas library to read excel
import pandas as pd

# define path for evoking excel file
path = r"C:\Users\Sony\Desktop\driver_info.xlsx"

# evoking 'اطلاعات سامانه جامع' sheet from excel file
driver_info_sheet1=pd.read_excel(path, sheet_name=)
driver_info_sheet1.head(n=3)
```

```python
# finding columns having None value
display(driver_info_sheet1.isna().any())
```

```python
# changing 'پلاک' format

driver_info_sheet1['پلاک']=driver_info_sheet1['پلاک'].str.replace('|','-')

# 
driver_info_sheet1['پلاک']=driver_info_sheet1['پلاک'].str.strip()

# removing space in 'پلاک' column
driver_info_sheet1['پلاک']=driver_info_sheet1['پلاک'].str.replace(' ','')
driver_info_sheet1.head(n=3)
```

```python
driver_info_sheet2=pd.read_excel(path, sheet_name='اطلاعات راهنمایی رانندگی')

# removing 'ایران' in 'پلاک' column
driver_info_sheet2['پلاک']=driver_info_sheet2['پلاک'].str.replace('ايران','')
driver_info_sheet2.head(n=3)
```

```python
frames=[driver_info_sheet1,driver_info_sheet2]

# combining 'اطلاعات سامانه جامع' & 'اطلاعات راهنمایی رانندگی' sheet
combined_driver_info=pd.concat(frames)

# adding 'ایران' to any 'پلاک' column's context
combined_driver_info['پلاک']=combined_driver_info['پلاک'].apply(lambda x: str(f"ایران{x}"))
combined_driver_info=combined_driver_info.fillna("police")
combined_driver_info.head(n=3)
```

```python
import numpy as np
matrix_driver=np.array(combined_driver_info)
```

```python
deleted_data=[]
for i in range(0,matrix_driver.shape[0]-1):
    for j in range(i+1,matrix_driver.shape[0]):
        if matrix_driver[i,0]==matrix_driver[j,0]:
            print(matrix_driver[i,0],matrix_driver[j,0])
            if matrix_driver[i,2]=="police":
                matrix_driver[i,2]=matrix_driver[j,2]
            if matrix_driver[i,5]=='police':
                matrix_driver[i,5]=matrix_driver[j,5]
            if matrix_driver[i,8]=='police':
                matrix_driver[i,8]=matrix_driver[j,8]
            if matrix_driver[i,9]=='police':
                matrix_driver[i,9]=matrix_driver[j,9]
            if matrix_driver[i,10]=='police':
                matrix_driver[i,10]=matrix_driver[j,10]
            deleted_data.append(j)
        else:
            pass 
print(deleted_data)   
for i in deleted_data:
    matrix_driver[i]=None
Final_Driver_Data=pd.DataFrame(matrix_driver)
Final_Driver_Data.to_excel('Final_Driver_Data.xlsx',sheet_name='cleaned')
```
